```python
import numpy as np
import matplotlib.pyplot as plt
```

# Point source in 2D

Although I haven't quite figured out why, the Greens function in 2D is apparantely:

$G(r_0, r) = \frac{i}{4} H_0^{(1)} (k |r - r_0|) $

And its derivative is:

$ \frac{\partial G}{\partial r} = - \frac{i}{4} k H_1^{(1)} $

To find the derivative across the boundary (along the normal), we can use the chain-rule:

$ \frac{\partial G}{\partial n} = \frac{\partial G}{\partial r}\frac{\partial r}{\partial n} $

$\frac{\partial G}{\partial n}(r_0, r) -\frac{i}{4}  k \frac{\partial r}{\partial n}  H_0^{(1)} (k |r - r_0|) $

The derivative of $r$ with respect to $n$ is calculated by:

$ \frac{\partial r}{\partial n} = \frac{\mathbf{r} \cdot \mathbf{n}}{r} $

```python
x = np.arange(0.1, 100, 1)
y = np.arange(0.1, 100, 1)
xx, yy = np.meshgrid(x, y)

# Source position
rx, ry = 50, 50
ki = 0.5

from scipy.special import hankel1

R = np.sqrt((rx-xx)**2 + (ry-yy)**2)
Fh = 1j/4 * hankel1(0, ki*R) # slightly infinite at rx, ry

f, ax = plt.subplots()
ax.imshow(np.real(Fh), vmin=0, vmax=1/4)
```

# Plane wave

$ G(x_0, x) = A \exp(i \mathbf{k \cdot x}) $

```python
# A wave traveling in the x-direction.
A = 1.
Fp = A * np.exp(1j * ki * xx)

f, ax = plt.subplots()
ax.imshow(np.real(Fp), vmin=-1, vmax=1)
```

```python
# An oblique one:

kx = 0.5
ky = 0.5

Fp = A * np.exp(1j * (kx*xx + ky*yy))

f, ax = plt.subplots()
ax.imshow(np.real(Fp), vmin=-1, vmax=1)
```
